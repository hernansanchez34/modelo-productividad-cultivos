```python
#Lectura y visualización de .soil
import matplotlib.pyplot as plt

def leer_smtab_desde_txt(archivo):
    smtab = []
    sm0 = smfcf = smw = None
    leyendo_smtab = False

    with open(archivo, "r") as f:
        for line in f:
            if "SMTAB" in line:
                leyendo_smtab = True
                line = line.replace("SMTAB    =", "")
            elif leyendo_smtab and "=" in line and "SMTAB" not in line:
                leyendo_smtab = False

            if leyendo_smtab:
                line = line.split("!")[0]
                valores = [v.strip() for v in line.strip().split(",") if v.strip()]
                for i in range(0, len(valores)-1, 2):
                    try:
                        pf = float(valores[i])
                        humedad = float(valores[i+1])
                        smtab.append((pf, humedad))
                    except:
                        continue

            if "SM0" in line:
                sm0 = float(line.split("=")[1].split("!")[0].strip())
            if "SMFCF" in line:
                smfcf = float(line.split("=")[1].split("!")[0].strip())
            if "SMW" in line:
                smw = float(line.split("=")[1].split("!")[0].strip())

    return smtab, sm0, smfcf, smw

# Rutas de los tres archivos
path1 = "data/soil/ec1_convertido.txt"
path2 = "data/soil/ec2_convertido.txt"
path3 = "data/soil/ec3_convertido.txt"

# Leer los datos
smtab_ec1, sm0_1, smfcf_1, smw_1 = leer_smtab_desde_txt(path1)
smtab_ec2, sm0_2, smfcf_2, smw_2 = leer_smtab_desde_txt(path2)
smtab_ec3, sm0_3, smfcf_3, smw_3 = leer_smtab_desde_txt(path3)

# Separar pF y humedad
pF_1, h_1 = zip(*smtab_ec1)
pF_2, h_2 = zip(*smtab_ec2)
pF_3, h_3 = zip(*smtab_ec3)

# Crear figura con 3 subplots horizontales
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(18, 5), sharey=True)

# --- EC1 ---
ax1.plot(pF_1, h_1, 'o-', color='teal', label="EC1 (coarse)")
ax1.axhline(sm0_1, color='blue', linestyle='--', label='SM0')
ax1.axhline(smfcf_1, color='green', linestyle='--', label='SMFCF')
ax1.axhline(smw_1, color='red', linestyle='--', label='SMW')
ax1.invert_xaxis()
ax1.set_title("Curva de retención - EC1")
ax1.set_xlabel("pF")
ax1.set_ylabel("Contenido volumétrico")
ax1.grid(True)
ax1.legend()

# --- EC2 ---
ax2.plot(pF_2, h_2, 'o-', color='orange', label="EC2 (medium)")
ax2.axhline(sm0_2, color='blue', linestyle='--', label='SM0')
ax2.axhline(smfcf_2, color='green', linestyle='--', label='SMFCF')
ax2.axhline(smw_2, color='red', linestyle='--', label='SMW')
ax2.invert_xaxis()
ax2.set_title("Curva de retención - EC2")
ax2.set_xlabel("pF")
ax2.grid(True)
ax2.legend()

# --- EC3 ---
ax3.plot(pF_3, h_3, 's-', color='purple', label="EC3 (medium fine)")
ax3.axhline(sm0_3, color='blue', linestyle='--', label='SM0')
ax3.axhline(smfcf_3, color='green', linestyle='--', label='SMFCF')
ax3.axhline(smw_3, color='red', linestyle='--', label='SMW')
ax3.invert_xaxis()
ax3.set_title("Curva de retención - EC3")
ax3.set_xlabel("pF")
ax3.grid(True)
ax3.legend()

# Mostrar figura
plt.tight_layout()
plt.show()
```
